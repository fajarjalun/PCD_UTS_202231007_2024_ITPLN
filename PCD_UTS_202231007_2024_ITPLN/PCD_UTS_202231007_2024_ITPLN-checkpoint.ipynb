# IMPORT LIBRARY

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# MEMBACA GAMBAR

```python
img = cv2.imread('tulisan.jpg')
```

```python
img.shape
```

```python
[baris, kolom] = img.shape[:2]
```

# BGR TO RGB

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

# WARNA MERAH

```python
red_channel = img[:,:,0]
citra_merah = cv2.merge([red_channel, red_channel, red_channel])

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,25])
axs[1,0].imshow(citra_merah)
axs[1,1].hist(citra_merah.ravel(), 256, [0,25])
plt.show()
```

# WARNA HIJAU

```python
green_channel = img[:,:,1]
citra_hijau = cv2.merge([green_channel, green_channel, green_channel])

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,250])
axs[1,0].imshow(citra_hijau)
axs[1,1].hist(citra_hijau.ravel(), 256, [0,250])
plt.show()
```

# WARNA BIRU

```python
blue_channel = img[:,:,2]
citra_biru = cv2.merge([blue_channel, blue_channel, blue_channel])

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(), 256, [0,250])
axs[1,0].imshow(citra_biru)
axs[1,1].hist(citra_biru.ravel(), 256, [0,250])
plt.show()
```

# RANGKUMAN MEMISAHKAN WARNA

```python
alpha = 1.1
citra_kontras = np.zeros((baris, kolom, 3))

for x in range(baris):
    for y in range(kolom):
        gmx = img[x,y] * alpha
        citra_kontras[x,y]= gmx

citra_kontras = citra_kontras.astype(np.uint8)

red_channel = img[:,:,0]
red_image = cv2.merge([red_channel, red_channel, red_channel])

green_channel = img[:,:,1]
green_image = cv2.merge([green_channel, green_channel, green_channel])

blue_channel = img[:,:,2]
blue_image = cv2.merge([blue_channel, blue_channel, blue_channel])

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0, 0].set_title('Kontras')
axs[0,0].imshow(citra_kontras)
axs[0, 1].set_title('Merah')
axs[0,1].imshow(red_image)
axs[1, 0].set_title('Hijau')
axs[1,0].imshow(green_image)
axs[1, 1].set_title('Biru')
axs[1,1].imshow(blue_image)

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
fig.suptitle("Memisahkan Warna")
plt.show()
```

```python
tulisan_merah_biru = cv2.merge([red_channel, blue_channel, blue_channel])

# Mengganti warna tulisan menjadi putih dan latar belakang menjadi hitam
greyscale_red_blue = cv2.cvtColor(tulisan_merah_biru, cv2.COLOR_BGR2GRAY)

plt.imshow(greyscale_red_blue, cmap='gray')
plt.show()
```

# MEMBUAT CITRA NONE

```python
citra_none = np.zeros_like(img)
plt.imshow(citra_none)
```

# MENCARI AMBANG BATAS

```python
grayscale_biru = cv2.cvtColor(citra_biru, cv2.COLOR_BGR2GRAY)

# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_1, binary_biru = cv2.threshold(grayscale_biru, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _1)
```

```python
grayscale_hijau = cv2.cvtColor(citra_hijau, cv2.COLOR_BGR2GRAY)

# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_2, binary_hijau = cv2.threshold(grayscale_hijau, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _2)
```

```python
grayscale_merah = cv2.cvtColor(citra_merah, cv2.COLOR_BGR2GRAY)

# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_3, binary_merah = cv2.threshold(grayscale_merah, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _3)
```

```python
grayscale_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_4, binary_rgb = cv2.threshold(grayscale_rgb, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _4)
```

```python
grayscale_none = cv2.cvtColor(citra_none, cv2.COLOR_BGR2GRAY)

# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_5, binary_none = cv2.threshold(grayscale_none, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _5)
```

```python
# Menggunakan metode thresholding adaptif untuk menentukan batas ambang
_6, binary_red_blue = cv2.threshold(greyscale_red_blue, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("Nilai ambang batas:", _6)
```

```python
nama_citra = ["Biru", "Hijau", "Merah", "RGB", "None"]
threshold_values = [_1, _2, _3, _4, _5]

# Menggabungkan nama citra dengan nilai threshold menggunakan zip
nama_threshold = list(zip(nama_citra, threshold_values))

# Mengurutkan berdasarkan nilai threshold
sorted_nama_threshold = sorted(nama_threshold, key=lambda x: x[1])

print("Citra yang diurutkan berdasarkan nilai threshold:")
for citra, threshold in sorted_nama_threshold:
    print(f"{citra} -> Nilai ambang batas = {threshold}")
```

```python
# Invert citra
inverted_citra_biruuu = cv2.bitwise_not(binary_biru)
inverted_citra_hijau = cv2.bitwise_not(binary_hijau)
inverted_citra_merah =  cv2.bitwise_not(binary_merah)
inverted_citra_rgb = cv2.bitwise_not(binary_rgb)

# Membuat subplot untuk menampilkan citra dan histogramnya
fig, axs = plt.subplots(2,2,figsize=(15,5))

# Menampilkan citra dan histogram inverted_citra_kontras
axs[0,0].imshow(inverted_citra_hijau, cmap="gray")
axs[0,0].set_title('Inverted Citra Hijau')
axs[0,1].imshow(inverted_citra_merah, cmap="gray")
axs[0,1].set_title('Inverted Citra Merah')

# Menampilkan citra dan hiastogram inverted_citra_biru
axs[1,0].imshow(inverted_citra_biruuu, cmap="gray")
axs[1,0].set_title('Inverted Citra Biru')
axs[1,1].imshow(inverted_citra_rgb, cmap="gray")
axs[1,1].set_title('Inverted Citra RGB')

plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.show()
```
